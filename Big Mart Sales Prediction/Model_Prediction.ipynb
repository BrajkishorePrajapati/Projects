## Import required libraries

```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

## Load the dataset.

```
# LOading train dataset
train = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Projects/Big Mart sales/train_modified.csv')
test = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Projects/Big Mart sales/test_modified.csv')
submission = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Projects/Big Mart sales/SampleSubmission.txt')
```

* **Set the index of train and test dataset as Item_Identifier and Outlet_Identifier**

```
train.set_index(['Item_Identifier','Outlet_Identifier'],inplace = True)
test.set_index(['Item_Identifier','Outlet_Identifier'],inplace = True)
```

```
train.head()
```

```
test.head()
```

```
X = train.drop('Item_Outlet_Sales', axis = 1)
Y = train['Item_Outlet_Sales']
```

**Here is Dataframe contains independent features i.e input variable, whereas Y is Series contains dependent fearture i.e target variable.**

## **Split the train into train and dev dataset**


```
X_train, X_dev, y_train, y_dev = train_test_split(X,Y, test_size = 0.1, random_state = 0)
print('X_train','\n')
print(X_train)
print('*'*80)
print('X_dev','\n')
print(X_dev)
print('y_train','\n')
print(y_train)
print('*'*80)
print('y_dev','\n')
print(y_dev)
```

## **Model Building**

## **Linear Regression**

```
# make the object 
lr = LinearRegression(normalize = True)
# Fit or train the model
lr.fit(X_train,y_train)
y_pred = lr.predict(X_dev)
b_hat=lr.intercept_
print('Intercept value is :',b_hat)
print('*'*80)
w_hat = lr.coef_
print('Coeffient of multiple variables are :',w_hat)
print('*'*80)

# Mean squared error
mse_score = mean_squared_error(y_dev,y_pred)
print('Mean Squared error on development set is ',mse_score)

# R squared error
r_score_train = lr.score(X_train,y_train)
print('R Square of trainng score is :',r_score_train)
r_score_dev = lr.score(X_dev,y_dev)
print('R Square of development score is :',r_score_dev)

```

## **Decision Tree Regrressor**

```
dt = DecisionTreeRegressor(max_depth=15, min_samples_leaf= 100)
dt.fit(X_train,y_train)
y_pred_dt = dt.predict(X_dev)

# Mean squared error
mse_score = mean_squared_error(y_dev,y_pred_dt)
print('Mean Squared error on development set is ',mse_score)

# R squared error
r_score_train = dt.score(X_train,y_train)
print('R Square of trainng score is :',r_score_train)
r_score_dev = dt.score(X_dev,y_dev)
print('R Square of development score is :',r_score_dev)
```

```
#y_prediction = lr.predict(test)
y_prediction = dt.predict(test)
```

## **Modifying submission File**

```
submission['Item_Outlet_Sales'] = abs(y_prediction)
submission.head()
```
